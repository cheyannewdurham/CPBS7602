```python
import pandas as pd
```

```python
# data file paths
data_path = "/home/cheyanne/github_repos/CPBS7602/assignment01/data/tpm_5kgenes_10tissues.csv"
```

```python
# load data
data = pd.read_csv(data_path, index_col=0)
```

```python
print(data.head())
```

```python
# Replace 'GTEX' with the corresponding 'tissue_id' values in the index
data.index = data['tissue_id'] + '-' + data.index.str.split('-').str[1:].str.join('-')
print(data.head())
```

```python
# delete the first hyphen and everyting after it
data.index = data.index.str.split('-').str[0]
print(data.head())
```

```python
# remove the 'tissue_id' column
data = data.drop(columns='tissue_id')
print(data.head())
```

```python
# save the data
data.to_csv("/home/cheyanne/github_repos/CPBS7602/assignment01/data/tpm_5kgenes_10tissues_formatted.csv")
```
