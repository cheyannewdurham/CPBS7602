```python
import numpy as np
import pandas as pd
```

```python
# define pickle and metadata and output filepaths
tpm_pkl_fp = "/home/cheyanne/github_repos/CPBS7602/assignment01/data/GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_tpm-top5k_var_genes.pkl"
md_fp = "/home/cheyanne/github_repos/CPBS7602/assignment01/data/metadata.txt"
tpm_out_fp = "/home/cheyanne/github_repos/CPBS7602/assignment01/data/tpm_5kgenes_10tissues.csv"
```

```python
# load pickle tpm data
tpm_data = pd.read_pickle(tpm_pkl_fp)
print(tpm_data.head)
```

```python
# load meta data
md = pd.read_csv(md_fp, sep="\t", index_col=0)
print(md)
```

```python
# count the unique tissues in the smtsd column
tissue_counts = md['SMTS'].value_counts()
print(tissue_counts)
```

```python
# report the top 10 tissues in the tissue_counts
top_10_tissues = tissue_counts.head(10)
print(top_10_tissues)
```

```python
# remove columns from the md dataframe besides the Sindex and SMTS columns
md = md[['SMTS']]
print(md)
```

```python
# Filter the md data to only include samples from the top 10 tissues
top_10_tissues = top_10_tissues.index
md_filtered = md[md['SMTS'].isin(top_10_tissues)]
print(md_filtered)
```

```python
# create a key that lists the sample IDs for each tissue
tissue_sample_ids = {}
for tissue in top_10_tissues:
    tissue_sample_ids[tissue] = md_filtered[md_filtered['SMTS'] == tissue].index
print(tissue_sample_ids)
```

```python
# transpose the tpm_data
tpm_data = tpm_data.T
print(tpm_data)
```

```python
# Extract 'Name' column values from md_filtered and cross reference to tpm_data SAMPID column
sampid = md_filtered.index
tpm_data = tpm_data[tpm_data.index.isin(sampid)]
print(tpm_data)
```

```python
# search the tissue_sample_ids dictionary for the tissue that the sample belongs to and add it to the tpm_data
tissue = []
for sample in tpm_data.index:
    for key, value in tissue_sample_ids.items():
        if sample in value:
            tissue.append(key)
            break
tpm_data['tissue_id'] = tissue
print(tpm_data)
```

```python
tpm_data.to_csv(tpm_out_fp)
```
